## Query
__Compare review counts vs. star rating__

_If you have more reviews, does your star rating go up?_

```python
# sort the data by review count 
vegan_brklyn_df.sort_values(by='review_count', inplace=True)
```

```python
vegan_brklyn_df.head()
```

```python
# plot with matplotlib
# get x and y axis
x = vegan_brklyn_df['review_count']
y = vegan_brklyn_df['rating']

# Use plot() function to create a plot using above values on both x and y coordinates
plt.figure(figsize=(12,10))
plt.bar(x, y)
# plt.scatter(x, y, s=len(vegan_brklyn_df['price']))
plt.xlabel('Review Counts')
plt.ylabel('Rating')
plt.title('Review Count/Rating for 30 Vegan Rest in Brooklyn', fontsize=16)
plt.show()
plt.savefig('./HW/Yelp_API.png', dpi=300)
```
